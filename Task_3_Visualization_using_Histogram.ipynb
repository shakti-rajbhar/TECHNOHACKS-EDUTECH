Create a histogram or bar chart to visualize the distribution of data in a dataset

```python
#importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
data = iris.data
sepal_length = data[:,0]
```

```python
plt.figure(figsize=(8,6))
sns.histplot(sepal_length , bins=20, kde= False, color='green' , edgecolor='black')

plt.xlabel('Sepal Length(cm)')
plt.ylabel('Frequecy')
plt.title('Hisogram of Sepal Length in Iris Dataset')
plt.show()
```
