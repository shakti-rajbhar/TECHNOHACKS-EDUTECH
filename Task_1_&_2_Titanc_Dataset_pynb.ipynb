*   Clean a dataset by removing missing values and outliers
*   Calculate summary statistics (mean, median, mode, standard deviation)

```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading the dataset
df = pd.read_csv('/content/train.csv')
df.head(10)
```

```python
df.tail(10)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
missingValues = (df.isnull().sum()/len(df))*100
print(missingValues)
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df.dropna(how="all")
```

```python
df.dropna(subset=['Embarked'] ,inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Cabin"].value_counts()
```

```python
df["Cabin"].mode()
```

```python
df['Cabin'].fillna(df['Cabin'].mode()[2],inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.boxplot(x=df["Age"],showfliers=True)
```

```python
Q1= df.Age.quantile(0.25)
Q3= df.Age.quantile(0.75)
Q1,Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_limit= Q1-1.5*IQR
upper_limit= Q3+1.5*IQR
lower_limit,upper_limit
```

```python
df[(df.Age<lower_limit)|(df.Age>upper_limit)]
```

```python
no_outlier=df[(df.Age>=lower_limit)&(df.Age<=upper_limit)]
no_outlier
```

```python
sns.boxplot(x=no_outlier["Age"],showfliers=True)
```

```python
df["Age"].mean()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.boxplot(x=df["Fare"],showfliers=True)
```

```python
Q1=df.Fare.quantile(0.25)
Q3=df.Fare.quantile(0.75)
IQR=Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
no_outlier_fare=df[(df.Fare>=lower_limit)&(df.Fare<=upper_limit)]
no_outlier_fare
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```
