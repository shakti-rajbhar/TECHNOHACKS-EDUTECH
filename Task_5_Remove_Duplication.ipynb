Identify and remove duplicate values in a dataset.

**Task 5: Data Cleaning Iris Dataset**

```python
#importing libraries
import pandas as pd
```

```python
#loading dataset
iris = pd.read_csv('/content/Iris.csv')
iris.head()
```

```python
#data summary
iris.info()
```

```python
#descriptive stats
iris.describe()
```

```python
#duplicate rows identification
duplicates = iris[iris.duplicated(subset=['SepalLengthCm', 'SepalWidthCm', 'PetalWidthCm' , 'PetalLengthCm', 'Species'])]
duplicates
```

```python
#removing duplicate values
unique_data = iris.drop_duplicates(subset=['SepalLengthCm', 'SepalWidthCm', 'PetalWidthCm' , 'PetalLengthCm', 'Species'])
print("\nNumber of Duplicates Removed:", len(duplicates))
unique_data
```

```python
#save data
Irisoutput_csv = "unique_iris_data.csv"
unique_data.to_csv(Irisoutput_csv, index = False)
```
